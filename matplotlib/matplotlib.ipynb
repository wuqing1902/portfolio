# Introduction 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#### Explanation 
This step imports the required libraries:

- NumPy → used for numerical data creation and manipulation
- Matplotlib → used for data visualization
- Pandas → used for structured data handling (DataFrame format)

This is the foundation required before any plotting begins.

<br><br><br>

# Create a Consistent Dataset (Student Performance)

```python
np.random.seed(42)

students = np.array(["A", "B", "C", "D", "E", "F", "G", "H"])
study_hours = np.array([2, 3, 4, 5, 6, 7, 8, 9])
exam_scores = np.array([55, 60, 65, 70, 75, 80, 90, 95])
```

#### Explanation 
This step generates a consistent dataset representing:

- students
- study hours
- exam scores

The purpose is to simulate a real-world scenario where we analyze student performance. This dataset will be reused across all visualizations to maintain consistency.

<br><br><br>

# Line Plot — Performance Trend

```python
plt.plot(study_hours, exam_scores, marker='o', linestyle='--', color='blue')

plt.title("Study Hours vs Exam Scores")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.grid(True)

plt.show()
```

#### Explanation 
A line plot is used to show how exam scores change as study hours increase.

It helps visualize:

- trends
- progression
- relationship over ordered data

This is useful for understanding whether performance improves with study time.

<br><br><br>

# Scatter Plot — Relationship Analysis

```python
plt.scatter(study_hours, exam_scores, color='green', s=100, alpha=0.7)

plt.title("Correlation Between Study Hours and Scores")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")

plt.show()
```

#### Explanation 
A scatter plot shows the relationship between study hours and exam scores.

It helps identify:

- correlation (positive or negative)
- clustering patterns
- outliers

This is one of the most important plots in data analysis.

<br><br><br>

# Bar Chart — Student Comparison

```python
plt.bar(students, exam_scores, color='skyblue')

plt.title("Exam Scores by Student")
plt.xlabel("Students")
plt.ylabel("Scores")

plt.show()
```

#### Explanation 
A bar chart compares exam scores across different students.

It is useful for:

- ranking individuals
- comparing categories
- showing discrete differences

Each bar represents one student’s performance.

<br><br><br>

Horizontal Bar Chart — Ranking View

```python
plt.barh(students, exam_scores, color='orange')

plt.title("Student Ranking (Horizontal View)")
plt.xlabel("Scores")
plt.ylabel("Students")

plt.show()
```

#### Explanation 
This is the same comparison as the bar chart but displayed horizontally.

It improves readability when:

- labels are long
- ranking comparison is needed

It is commonly used in dashboards.

<br><br><br>

# Histogram — Score Distribution

```python
plt.hist(exam_scores, bins=5, color='purple', edgecolor='black', alpha=0.7)

plt.title("Distribution of Exam Scores")
plt.xlabel("Score Range")
plt.ylabel("Number of Students")

plt.show()
```

#### Explanation 
A histogram shows how exam scores are distributed.

It helps answer:

- Are most students scoring high or low?
- Is the distribution balanced or skewed?
- How spread out are the scores?

It groups values into ranges (bins).

<br><br><br>

# Pie Chart — Performance Categories

```python
labels = ["High", "Mid", "Low"]

high = len(exam_scores[exam_scores >= 80])
mid = len(exam_scores[(exam_scores >= 60) & (exam_scores < 80)])
low = len(exam_scores[exam_scores < 60])

sizes = [high, mid, low]

plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    explode=(0.1, 0, 0),
    shadow=True
)

plt.title("Student Performance Categories")
plt.show()
```

#### Explanation 
This chart divides students into performance groups:

- High performers
- Mid performers
- Low performers

It shows the proportion of each group in percentage form, helping understand the overall class structure.

<br><br><br>

# Multi-Line Comparison (Advanced Insight)

```python
plt.plot(study_hours, exam_scores, marker='o', label="Scores")
plt.plot(study_hours, study_hours * 10, marker='s', label="Ideal Growth Line")

plt.title("Actual vs Ideal Performance Trend")
plt.xlabel("Study Hours")
plt.ylabel("Value")
plt.legend()
plt.grid(True)

plt.show()
```

#### Explanation 
This plot compares:

- actual student performance (scores)
- an ideal growth trend (reference line)

It helps evaluate:

- how close students are to expected performance
- whether performance is above or below expectation

<br><br><br>

# Subplots Dashboard (FULL SUMMARY VIEW)

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Line plot
axes[0, 0].plot(study_hours, exam_scores, marker='o')
axes[0, 0].set_title("Trend")

# Bar chart
axes[0, 1].bar(students, exam_scores, color='skyblue')
axes[0, 1].set_title("Scores")

# Scatter plot
axes[1, 0].scatter(study_hours, exam_scores, color='green')
axes[1, 0].set_title("Correlation")

# Histogram
axes[1, 1].hist(exam_scores, bins=5, color='purple', edgecolor='black')
axes[1, 1].set_title("Distribution")

plt.tight_layout()
plt.show()
```

#### Explanation 
Subplots allow multiple charts to be displayed in one figure.

This creates a mini dashboard showing:

- trend (line plot)
- comparison (bar chart)
- relationship (scatter plot)
- distribution (histogram)

It is used for summary reporting and data dashboards.

<br><br><br>
