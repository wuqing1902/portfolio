# Copy Safely

```python
print("ORIGINAL df (before copy):")
print(df)

print("\n" + "-"*50 + "\n")

print("Create independent copy: df_copy = df.copy()")
df_copy = df.copy()

print('Modify ONLY the copy: df_copy["Salary"] = df_copy["Salary"] * 2')
df_copy["Salary"] = df_copy["Salary"] * 2

print("Show both DataFrames")
print("ORIGINAL df (after modifying copy):")
print(df)
print("\n")

print("COPIED df_copy (modified):")
print(df_copy)

print("\n" + "-"*50 + "\n")

print("Proof using memory location (object identity)")
print("Memory ID of df:", id(df))
print("Memory ID of df_copy:", id(df_copy))
```

#### Explanation
Creates an independent copy of the dataset to avoid accidental modifications.
