### Basic Language Fundamentals

##### Variables
- A variable stores data in memory.
- Python automatically determines the variable's data type.
- Use **snake_case** for better readability and to follow the PEP 8 style guide.
- Choose meaningful names that clearly describe the data they store.
- Variable names must follow Python's naming rules.

    - May contain letters, numbers, and underscores 
    - Cannot contain spaces 
    - Cannot contain special characters

1. Accessing characters - Characters in a string are accessed using **zero-based indexing**.

3. String Concatenation - Strings can be combined using the `+` operator.
4. String Slicing - Slicing extracts a portion of a string using the syntax `[start:end:step]`.
5. Common string methods

    - `lower()` - Converts all characters to lowercase.

    - `title()` - Converts the first character of each word in the string to uppercase.
    - `replace()` - Replaces part of a string with another value.
    - `strip()` - Removes leading and trailing whitespace.
    - `split()` - Splits a string into a list. 

````python
print("Accessing Characters:")
text = "Python"
print('text = "Python"')
print("text[0]: ", text[0])
print("text[2]: ", text[2])
print("text[-1]: ", text[-1])

print("\n")

print("Escaping Characters:")
print("He said, \"Hello!\"")
print('It\'s Python.')
print("Line 1\nLine 2")

print("\n")

print("String Concatenation:")
print('first_name = "John"')
first_name = "John"
print('last_name = "Doe"')
last_name = "Doe"
print('full_name = first_name + " " + last_name')
full_name = first_name + " " + last_name
print('full_name: ', full_name)

print("\n")

print("String Slicing:")
text = "Python"
print('text = "Python"')
print("text[0:3]: ", text[0:3])
print("text[2:5]: ", text[2:5])
print("text[:4]: ", text[:4])
print("text[2:5]: ", text[2:5])
print("text[::-1]: ", text[::-1])

print("\n")

print("Common string methods:")
text = "    Hello Python    "
print('text = "    Hello Python    "')
print("text.upper(): ", text.upper())
print("text.lower(): ", text.lower())
print("text.capitalize(): ", text.capitalize())
print("text.title(): ", text.title())
print("text.replace('P', 'p'): ", text.replace('P', 'p'))
print("print(text.strip()): ", text.strip())
print("text.split(): ", text.split())
````

### Scope
Scope determines where a variable can be accessed within a program. Python follows the LEGB Rule to resolve variable names until it finds a match.
```
LEGB = Local → Enclosing → Global → Built-in
```

##### Local Scope 
A local variable is defined inside a function and can only be accessed within that function.
```
def greet():
    message = "Hello, Python!"
    print(message)
greet()

# print(message)   # Uncomment this will returned an NameError: name 'message' is not defined. 
```


##### Global Scope
A global variable is defined outside any function and can be accessed throughout the program.
```
language = "Python"
def display():
    print(language)
display()
print(language)
```


##### Modifying a Global Variable
To modify a global variable inside a function, use the global keyword.
```
counter = 0
def increment():
    global counter      # Without this will returned an UnboundLocalError: local variable 'counter' referenced before assignment
    counter += 1

increment()
print(counter)
```


##### Enclosing Scope
An enclosing scope exists when a function is defined inside another function.
```
def outer():
    message = "Hello"
    def inner():
        print(message)
    inner()
outer()
```

### Python Lists
A list is an ordered, mutable collection that can store multiple items of different data types. Lists are created using square brackets [].

1. Creting List - Using syntax: `list_name = [item1, item2, item3]`
2. Accessing List Elements (Indexing) - Lists use zero-based indexing, meaning the first element has an index of 0.
3. Negative Indexing - Negative indexes count from the end of the list.
4. List Mutability - Lists are mutable, meaning their elements can be modified after creation.
5. List Slicing - Slicing extracts a portion of a list using the syntax [start:end:step].
7. Common List Methods: 
    - append() - Adds an element to the end of the list.
    - insert() - Inserts an element at a specified index.
    - extend() - Adds elements from another iterable.
    - remove() - Removes the first matching element.
    - pop() - Removes and returns an element.
    - clear() - Removes all elements.
    - sort() - Sorts the list.
    - reverse() - Reverses the list order.
    - count() - Counts occurrences of a value.
    - index() - Returns the first matching index.
    - copy() - Creates a shallow copy of the list.
    - len() - Returns the number of elements.

````python
# Creating Lists
fruits = ["apple", "banana", "orange"]

print("Original List:")
print(fruits)

# Accessing List Elements (Indexing)
print("\nAccessing Elements:")
print(f"First item: {fruits[0]}")
print(f"Second item: {fruits[1]}")

# Negative Indexing
print("\nNegative Indexing:")
print(f"Last item: {fruits[-1]}")
print(f"Second last item: {fruits[-2]}")

# List Mutability
print("\nList Mutability:")
fruits[1] = "grape"
print(f"Modified List: {fruits}")

# List Slicing
numbers = [10, 20, 30, 40, 50]
print("\nList Slicing:")
print(f"numbers[1:4] = {numbers[1:4]}")
print(f"numbers[:3] = {numbers[:3]}")
print(f"numbers[2:] = {numbers[2:]}")
print(f"numbers[::-1] = {numbers[::-1]}")

# append()
fruits.append("mango")
print("\nappend():")
print(fruits)

# insert()
fruits.insert(1, "kiwi")
print("\ninsert():")
print(fruits)

# extend()
fruits.extend(["pear", "watermelon"])
print("\nextend():")
print(fruits)

# remove()
fruits.remove("grape")
print("\nremove():")
print(fruits)

# pop()
removed_item = fruits.pop()
print("\npop():")
print(f"Removed Item: {removed_item}")
print(fruits)

# count()
numbers = [1, 2, 2, 3, 2]

print("\ncount():")
print(f"Occurrences of 2: {numbers.count(2)}")

# index()
print("\nindex():")
print(f"Index of 3: {numbers.index(3)}")

# copy()
numbers_copy = numbers.copy()
print("\ncopy():")
print(numbers_copy)

# len()
print("\nlen():")
print(f"Length of numbers: {len(numbers)}")

# sort()
unsorted = [5, 1, 8, 3, 2]
unsorted.sort()
print("\nsort():")
print(unsorted)

# reverse()
unsorted.reverse()
print("\nreverse():")
print(unsorted)

# clear()
temp = [1, 2, 3]
temp.clear()
print("\nclear():")
print(temp)
````

### Python Tuples 
A tuple is an ordered, immutable collection that can store multiple items of different data types. Tuples are created using parentheses ().

1. Creating Tuples - Using syntax: `tuple_name = (item1, item2, item3)`
2. Tuple Indexing
3. Tuple Slicing
4. Tuple Unpacking - Tuple unpacking assigns tuple elements to multiple variables.
5. Ignoring Values During Unpacking - Use _ to ignore values you don't need.
6. Tuple Immutability - Tuple elements cannot be modified.
7. Common Tuple Methods: 
    - count() - Returns the number of occurrences.
    - index() - Returns the index of the first matching element.

````python
# Creating Tuples
person = ("Alice", 25, "Engineer")
print("Original Tuple:")
print(person)

# Tuple Indexing
print("\nTuple Indexing:")
print(f"Name: {person[0]}")
print(f"Age: {person[1]}")
print(f"Job: {person[2]}")

# Tuple Slicing
print("\nTuple Slicing:")
print(person[0:2])
print(person[::-1])

# Tuple Unpacking
name, age, job = person
print("\nTuple Unpacking:")
print(name)
print(age)
print(job)

# Ignoring Values During Unpacking
name, _, job = person
print("\nIgnoring Values During Unpacking:")
print(name)
print(job)

# Tuple Immutability
print("\nTuple Immutability:")
try:
    person[0] = "Bob"
except TypeError as e:
    print(e)

# count()
numbers = (1, 2, 2, 3, 2)
print("\ncount():")
print(numbers.count(2))

# index()
print("\nindex():")
print(numbers.index(3))
````

### Loops
A loop repeatedly executes a block of code until a specified condition is met or all items in a collection have been processed.

1. for Loop - Iterates over each item in an iterable.
2. while Loop - Repeats while a condition remains True.
3. break - Terminates the loop immediately.
4. continue - Skips the current iteration.
5. pass - Placeholder statement that performs no action.
6. range() - Generates a sequence of numbers.
7. enumerate() - Returns both the index and value during iteration.
8. zip() - Combines multiple iterables element by element.

````python
# For Loop 
students = ["Alice", "Bob", "Charlie"]
for student in students:
    print(student)
print("\n")


# While Loop
count = 1
while count <= 3:
    print(count)
    count += 1
print("\n")


# Break
for number in range(5):
    if number == 3:
        break
    print(number)
print("\n")


# Continue
for number in range(5):
    if number == 2:
        continue
    print(number)
print("\n")


# Pass
for number in range(3):
    if number == 1:
        pass
    print(number)
print("\n")


# Enumerate
for index, student in enumerate(students, start=1):
    print(index, student)
print("\n")


# Zip
scores = [85, 90, 95]
for student, score in zip(students, scores):
    print(student, score)
````

### Functional Programming
Functional programming is a programming style that emphasizes using functions to transform data rather than modifying it directly. Python provides several built-in tools that make data processing concise and expressive.

1. List Comprehension - Creates a new list from an iterable.
2. filter() - Selects elements that satisfy a condition.
3. map() - Applies a function to every element.
4. sum() - Calculates the total of numeric values.
5. Lambda - Creates a short anonymous function.

````python
# List Comprehension 
numbers = [1, 2, 3, 4, 5]
squares = [number ** 2 for number in numbers]
print(squares)
print("\n")


# Filter
even = list(filter(lambda number: number % 2 == 0, numbers))
print(even)
print("\n")


# Map
doubled = list(map(lambda number: number * 2, numbers))
print(doubled)
print("\n")


# Sum
print(sum(numbers))
print("\n")


# Lambda
multiply = lambda a, b: a * b
print(multiply(6, 7))
print("\n")
````

### Dictionaries
A dictionary is a mutable collection that stores data as key-value pairs. Each key must be unique, while values can be of any data type.

1. Creating Dictionary - Use syntax: 
```
dictionary_name = {
    key1: value1,
    key2: value2
}
```
or 
```
dictionary_name = dict(
    key1: value1,
    key2: value2
)
```
2. Accessing Dictionary Values (Bracket Notation) - Dictionary values can be accessed using their corresponding keys.
3. The get() Method - The get() method retrieves a value without raising an error if the key does not exist.
4. Adding and Updating Elements
```
dict['newKey'] = 'newValue'     # Add
dict['updateKey'] = 'updateValue'       # Update
```
5. Removing Elements 
    - pop() - removes a key and returns its value 
    - popitem() - removes and returns the last inserted key-value pair
    - del - remove certain attribute inside dictionary
    - clear() - delete all the things inside dictionary
6. Common Dictionary Methods 
    - keys() - returns all the keys 
    - values() - returns all the values 
    - items() - returns key-value pairs as tuples 
    - update() - updates existing keys or adds new one 
    - copy() - creates a shallow copy 
    

````python
# Creating Dictionary
student = {
    "name": "Alice",
    "age": 20,
    "course": "Cybersecurity"
}
print("Original Dictionary:")
print(student)
print("\n")


# Accessing values
print("\nAccessing Values:")
print(student["name"])
print(student.get("age"))
print(student.get("email", "Not Found"))
print("\n")


# Adding / Updating
student["age"] = 21
student["university"] = "Harvard University"
print("\nAfter Update/Add:")
print(student)
print("\n")


# Removing elements
removed_age = student.pop("age")
print("\npop('age'):", removed_age)
temp_student = student.copy()
print("copy():", temp_student)
print("\n")


# popitem()
last_item = student.popitem()
print("popitem():", last_item)
print("\n")


# Dictionary methods
print("\nKeys:", student.keys())
print("Values:", student.values())
print("Items:", student.items())
print("\n")


# update()
student.update({"country": "Malaysia", "course": "Python"})
print("\nAfter update():")
print(student)
print("\n")


# Looping Dictionary
print("\nLooping Keys:")
for key in student:
    print(key)

print("\nLooping Values:")
for value in student.values():
    print(value)

print("\nLooping Items:")
for key, value in student.items():
    print(f"{key}: {value}")
print("\n")


# clear()
backup = student.copy()
student.clear()
print("\nAfter clear():", student)
print("Backup:", backup)
````

### Sets
A set is an unordered, mutable collection of unique elements. Duplicate values are automatically removed.

1. Create Set - Use syntax: `set_name = {item1, item2, item3}`
2. Adding Elements
    - add() - Adds an element.
3. Removing Elements 
- remove() - Removes an element (raises an error if absent).
- discard() - Removes an element safely.
- pop() - Removes and returns an arbitrary element.
- clear() - Removes all elements.
4. Mathematical Set Operations 
- union() - Combines unique elements from multiple sets.
- intersection() - Returns common elements.
- difference() - Returns elements unique to the first set.
- symmetric_difference() - Returns elements found in only one set.
- issubset() - Checks if one set is a subset of another.
- issuperset() - Checks if one set is a superset of another.
- isdisjoint() - Checks if two sets share no elements.

````python
# Creating Sets
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}
print("Set A:", set_a)
print("Set B:", set_b)
print("\n")


# add()
set_a.add(10)
print("\nAfter add(10):", set_a)
print("\n")


# remove() vs discard()
set_a.remove(10)
print("After remove(10):", set_a)
set_a.discard(99)  # no error if not found
print("After discard(99):", set_a)
print("\n")


# pop()
removed_item = set_a.pop()
print("\npop(): removed ->", removed_item)
print("Set A now:", set_a)
print("\n")


# clear()
temp_set = {1, 2, 3}
temp_set.clear()
print("\nclear():", temp_set)
print("\n")


# Set Operations
print("\nUnion:")
print(set_a | set_b)

print("\nIntersection:")
print(set_a & set_b)

print("\nDifference (A - B):")
print(set_a - set_b)

print("\nSymmetric Difference:")
print(set_a ^ set_b)
print("\n")


# Subset / Superset / Disjoint
x = {1, 2}
y = {1, 2, 3}

print("\nSubset:", x.issubset(y))
print("Superset:", y.issuperset(x))

z = {7, 8}
print("Disjoint:", x.isdisjoint(z))
````

### Object-Oriented Programming (OOP)
Object-Oriented Programming (OOP) is a programming paradigm that organizes code into objects, which combine data (attributes) and behavior (methods). OOP promotes code reusability, modularity, and maintainability, making it widely used in modern software development.

**Concepts:**
- Class – A blueprint or template for creating objects.
- Object – An instance of a class with its own data.
- Attribute – A variable that stores data for an object.
- Method – A function defined inside a class that describes an object's behavior.

### Classes and Objects 
##### Class
A class is a blueprint that defines the attributes and methods shared by its objects. Creating a class alone does not allocate memory for object data; objects must be instantiated from the class.

##### Object
An object is an instance of a class. Each object has its own attributes while sharing the methods defined by the class.

##### Attributes
An attribute is a variable associated with a class or an object.
- Instance Attributes – Unique to each object.
    - Instance attributes belong to individual objects. Each object can have different values.

### Dunder Methods 
Dunder methods (short for double underscore methods) are special methods that allow your objects to integrate with Python's built-in operations.

**Examples include:**

- __init__() → Object initialization
- __str__() → The __str__() method defines the user-friendly string representation of an object. It is automatically called by print() or str().
- __repr__() → The __repr__() method provides the official or developer-friendly representation of an object. It is mainly used for debugging.
- __len__() → The __len__() method defines the behavior of the built-in len() function for your objects
- __eq__() → The __eq__() method customizes the behavior of the == operator.
- __add__() → The __add__() method customizes the + operator.

````python
class Student:
    school = "ABC University"

    def __init__(self, student_id, name, scores):
        self.student_id = student_id
        self.name = name
        self.scores = scores

    def average_score(self):
        return sum(self.scores) / len(self.scores)

    def __str__(self):
        return f"{self.name} ({self.student_id})"

    def __repr__(self):
        return f"Student('{self.student_id}', '{self.name}')"

    def __len__(self):
        return len(self.scores)

    def __eq__(self, other):
        return self.student_id == other.student_id


student1 = Student("S001", "Alice", [80, 90, 85])
student2 = Student("S001", "Alice", [75, 88])
student3 = Student("S002", "Bob", [92, 90])

print("=== __str__() ===")
print(student1)

print("\n=== __repr__() ===")
print(repr(student1))

print("\n=== Instance Method ===")
print(student1.average_score())

print("\n=== __len__() ===")
print(len(student1))

print("\n=== __eq__() ===")
print(student1 == student2)
print(student1 == student3)
````

### OOP Principles 
Object-Oriented Programming is built upon four fundamental principles:

1. Encapsulation - Protects an object's internal data and controls access to it.
2. Inheritance - Allows a class to inherit attributes and methods from another class.
3. Polymorphism - Enables the same interface to behave differently for different objects.
4. Abstraction - Hides implementation details and exposes only essential functionality.

##### Encapsulation
Encapsulation is the practice of bundling data (attributes) and methods (behavior) within a class while controlling access to the internal state of an object. This helps prevent accidental modification and improves maintainability.

**Access Levels in Python: **

1. Public - Accessible from anywhere. eg. name
2. Protected (Convention) - Intended for internal use within the class and subclasses. eg. _name
3. Private (Name Mangling) - Makes direct access more difficult by applying name mangling to reduce accidential access for private attributes. eg. __name

- Getters and Setters
    - Instead of accessing private data directly, use getter and setter methods to control how data is read and modified.

- Properties (@property)
    - Python provides the @property decorator, allowing methods to be accessed like attributes while still providing validation.

##### Inheritance
Inheritance allows one class (child/subclass) to inherit attributes and methods from another class (parent/superclass). This promotes code reuse and simplifies maintenance.

1. Single Inheritance 
    - Method Overriding 
        - A child class can replace a method inherited from its parent.
    - Using super() 
        - The super() function allows a child class to call methods from its parent class.
        
2. Multiple Inheritance 
    - A class can inherit from multiple parent classes.

##### Polymorphism
Polymorphism allows different objects to respond to the same method call in different ways.

##### Abstraction
Abstraction hides implementation details and exposes only the essential interface.

````python
class Employee:
    company = "OpenAI"

    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary

    @property
    def salary(self):
        return self.__salary

    @salary.setter
    def salary(self, value):

        if value > 0:
            self.__salary = value
        else:
            raise ValueError("Salary must be positive.")

    def work(self):
        print(f"{self.name} is working.")

class Manager(Employee):

    def work(self):
        print(f"{self.name} is managing a team.")

employee = Employee("Alice", 5000)
manager = Manager("Bob", 8000)

employee.work()
manager.work()

print(employee.salary)

employee.salary = 5500

print(employee.salary)
````

### Advanced OOP 

##### Abstract Base Classes (ABC)
An Abstract Base Class (ABC) defines a common interface for subclasses. It cannot be instantiated directly and may contain one or more abstract methods that subclasses are required to implement.

Think of an ABC as a blueprint that specifies what a subclass must do without dictating how it should do it.

##### Implementing an Abstract Class
Every subclass must implement all abstract methods.

**Why Use Abstract Classes?**
- Suppose you're building an application that supports different payment methods.
- Every payment method should provide a pay() method.
- Instead of relying on documentation, an abstract class enforces this requirement.
- Now every payment class is guaranteed to implement pay().

````python
# Why use abstract class example
from abc import ABC, abstractmethod

class Payment(ABC):
    @abstractmethod
    def pay(self, amount):
        pass

class CreditCard(Payment):
    def pay(self, amount):
        print(f"Paid ${amount} by credit card.")

class PayPal(Payment):
    def pay(self, amount):
        print(f"Paid ${amount} using PayPal.")
````

##### Composition vs Inheritance
1. Inheritance ("is-a" relationship)
    - Use inheritance when one object is a more specialized version of another.
    - A Dog is an Animal.
2. Composition ("has-a" relationship)
    - Use composition when one object contains or uses another object.
    - A Car has an Engine.

````python
# Composition Example
class Engine:
    def start(self):
        print("Engine started")

class Car:
    def __init__(self):
        self.engine = Engine()

    def drive(self):
        self.engine.start()
        print("Car is moving")

car = Car()
car.drive()
````

##### Method Resolution Order (MRO)
When multiple inheritance is used, Python follows the Method Resolution Order (MRO) to determine which method should be executed.

Based on the code below, Python searches in this order: 
D → B → C → A

````python
class A:
    def show(self):
        print("Class A")

class B(A):
    def show(self):
        print("Class B")

class C(A):
    def show(self):
        print("Class C")

class D(B, C):
    pass

obj = D()
obj.show()
print(D.__mro__)
````

### Searching Algorithm 
Searching determines whether a target value exists in a collection.

##### Linear Search 
A Linear Search examines each element one by one until the target is found or the collection ends.

- Best: O(1)
- Worst: O(n)

````python
def linear_search(numbers, target):
    for index, value in enumerate(numbers):
        if value == target:
            return index
    return -1
numbers = [15, 7, 30, 18, 25]
print(linear_search(numbers, 18))
print(linear_search(numbers, 99))
````

##### Binary Search
A Binary Search repeatedly divides a sorted collection in half until the target is found with a requirement that the data must already be sorted.

- Best: O(1)
- Worst: O(log n)

````python
def binary_search(numbers, target):
    left = 0
    right = len(numbers) - 1
    while left <= right:
        mid = (left + right) // 2
        if numbers[mid] == target:
            return mid
        elif numbers[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
numbers = [5, 10, 15, 20, 25, 30]
print(binary_search(numbers, 20))
print(binary_search(numbers, 99))
````

##### Merge Sort
Merge Sort uses the Divide and Conquer strategy.

Steps:
- Divide the list into halves.
- Sort each half recursively.
- Merge the sorted halves.

````python
def merge_sort(numbers):
    if len(numbers) <= 1:
        return numbers
    mid = len(numbers) // 2
    left = merge_sort(numbers[:mid])
    right = merge_sort(numbers[mid:])
    return merge(left, right)
def merge(left, right):
    result = []
    i = 0
    j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result
numbers = [8, 3, 5, 1, 9, 2]
print(merge_sort(numbers))
````

##### Tree Traversal
Traversal means visiting every node in a tree.

There are three common Depth-First Search (DFS) traversals.
1. Preorder - Root → Left → Right
2. Inorder - Left → Root → Right
3. Postorder - Left → Right → Root

````python
print("Preorder Traversal:")
def preorder(node):
    if node:
        print(node.value)
        preorder(node.left)
        preorder(node.right)
preorder(root)

print("\n")

print("Inorder Traversal:")
def inorder(node):
    if node:
        inorder(node.left)
        print(node.value)
        inorder(node.right)
inorder(root)

print("\n")

print("Postorder Traversal:")
def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.value)
postorder(root)
````

### Breadth-First Search (BFS)
BFS explores nodes level by level which uses a Queue.

````python
from collections import deque
graph = {
    "A": ["B", "C"],
    "B": ["D"],
    "C": ["E"],
    "D": [],
    "E": []
}

def bfs(start):
    visited = set()
    queue = deque([start])
    while queue:
        node = queue.popleft()
        if node not in visited:
            print(node)
            visited.add(node)
            queue.extend(graph[node])
            
bfs("A")
````

### Depth-First Search (DFS)
DFS explores one branch as deeply as possible before backtracking which uses recursion or a stack.

````python
graph = {
    "A": ["B", "C"],
    "B": ["D"],
    "C": [],
    "D": []
}

visited = set()

def dfs(node):
    if node in visited:
        return
    visited.add(node)
    print(node)
    for neighbor in graph[node]:
        dfs(neighbor)

dfs("A")
````

### Dynamic Programming (DP)
Dynamic Programming is an optimization technique that solves complex problems by:
- Breaking them into smaller subproblems.
- Solving each subproblem only once.
- Reusing previously computed results.

When Should You Use DP?
- A problem is a good candidate for Dynamic Programming if it has:
    - Overlapping Subproblems
    - Optimal Substructure

##### Naive Fibonacci 
- Time Complexity: O(2ⁿ)
- Space Complexity: O(n)

````python
## Naive Fibonacci
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)
print(fibonacci(10))
````

##### Memoization (Top-Down DP)
Memoization stores previously computed results so they are not recalculated.
- Time Complexity: O(n)
- Space Complexity: O(n)

````python
## Store inside memory list
memo = {}
def fibonacci(n):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    memo[n] = fibonacci(n - 1) + fibonacci(n - 2)
    return memo[n]
print(fibonacci(10))


## using python lru_cache
from functools import lru_cache

@lru_cache(maxsize=None)
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)
print(fibonacci(10))
````

##### Tabulation (Bottom-Up DP)
Instead of recursion, build the solution iteratively from the smallest subproblems.
- Time Complexity: O(n)
- Space Complexity: O(n)

````python
def fibonacci(n):
    if n <= 1:
        return n
    dp = [0, 1]
    for i in range(2, n + 1):
        dp.append(dp[i - 1] + dp[i - 2])
    return dp[n]
print(fibonacci(10))
````

##### Space Optimization
Only the previous two Fibonacci values are needed.
- Time Complexity: O(n)
- Space Complexity: O(1)

````python
def fibonacci(n):
    if n <= 1:
        return n
    
    first = 0
    second = 1

    for _ in range(2, n + 1):
        first, second = second, first + second

    return second
print(fibonacci(10))
````
